Here, instead of averaging the product $(d\theta^2\times dx^2)$ over particles $i$ and time $t$, we just stack up all the different products and look at a histogram. The distribution is generated by varying $t$ and $i$ in:

$\left\{(
    \left\|
        x_i(t+dt) - x_i(t)
    \right\|^2
    -
    \langle
        x_i(t+dt) - x_i(t)
    \rangle_t)
\times
    (\left\|
        \theta_i(t + \tau + dt) - \theta_i(t+\tau)
    \right\|^2
    -
    \langle
        \theta_i(t + \tau + dt) - \theta_i(t+\tau)
    \rangle_t)
\right\}$

where $\tau=\{0,10,100\}$ as shown in plots and $dt=10$frames

```python
h = hist(np.concatenate(atcstack), bins=1000)
xlim(-50,150)
xlabel('Correlation strength (arbitrary units: Area^2 * Radians^2)')
ylabel('Count, over time and particle')
title('tau = {}'.format(tau))
```
